```python
# prompt: create a new column Reviews_clean and convert the Reviews to lowercase

df['Reviews_clean'] = df['Reviews'].str.lower()

print("\nDataFrame with 'Reviews_clean' column:")
print(df.head())
```

```python
# prompt: Remove Punctuation marks from Reviews_clean

import re

# Function to remove punctuation
def remove_punctuation(text):
  if isinstance(text, str):
    return re.sub(r'[^\w\s]', '', text)
  return text

# Apply the function to the 'Reviews_clean' column
df['Reviews_clean'] = df['Reviews_clean'].apply(remove_punctuation)

print("\nDataFrame with 'Reviews_clean' after removing punctuation:")
print(df[['Reviews', 'Reviews_clean']].head())
```

```python
# prompt: Remove whitespaces from Reviews_clean

# Remove whitespaces from the 'Reviews_clean' column
df['Reviews_clean'] = df['Reviews_clean'].str.replace(r'\s+', '', regex=True)

print("\nDataFrame with 'Reviews_clean' after removing whitespaces:")
print(df[['Reviews', 'Reviews_clean']].head())
```

```python
# prompt: Tokenization in Reviews_clean

# Install NLTK if you haven't already
!pip install nltk
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

# Function for tokenization
def tokenize_text(text):
  if isinstance(text, str):
    return word_tokenize(text)
  return None

# Apply tokenization to the 'Reviews_clean' column
df['Reviews_tokens'] = df['Reviews_clean'].apply(tokenize_text)

print("\nDataFrame with 'Reviews_tokens' column:")
print(df[['Reviews', 'Reviews_clean', 'Reviews_tokens']].head())
```

```python
# prompt: Remove Stop words from Reviews_clean

nltk.download('stopwords')
from nltk.corpus import stopwords

# Get the English stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stopwords(tokens):
  if tokens is not None:
    return [word for word in tokens if word not in stop_words]
  return None

# Apply stop word removal to the 'Reviews_tokens' column
df['Reviews_cleaned_stopwords'] = df['Reviews_tokens'].apply(remove_stopwords)

print("\nDataFrame with 'Reviews_cleaned_stopwords' column:")
print(df[['Reviews', 'Reviews_clean', 'Reviews_tokens', 'Reviews_cleaned_stopwords']].head())
```

```python
# prompt: Lemmatization in Reviews_clean

nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer

# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Function for lemmatization
def lemmatize_tokens(tokens):
  if tokens is not None:
    return [lemmatizer.lemmatize(word) for word in tokens]
  return None

# Apply lemmatization to the 'Reviews_cleaned_stopwords' column
df['Reviews_lemmatized'] = df['Reviews_cleaned_stopwords'].apply(lemmatize_tokens)

print("\nDataFrame with 'Reviews_lemmatized' column:")
print(df[['Reviews', 'Reviews_cleaned_stopwords', 'Reviews_lemmatized']].head())
```

```python
# prompt: Count of most Frequent words

from collections import Counter

# Flatten the list of lists into a single list of words
all_words = [word for tokens_list in df['Reviews_lemmatized'].dropna() for word in tokens_list]

# Count the frequency of each word
word_counts = Counter(all_words)

# Get the 20 most common words
most_common_words = word_counts.most_common(20)

print("\n20 Most Common Words:")
for word, count in most_common_words:
  print(f"{word}: {count}")
```

```python
# prompt: generate Word cloud

!pip install wordcloud
from wordcloud import WordCloud

# Join all the lemmatized words into a single string
all_words_string = " ".join(all_words)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, random_state=21, max_font_size=110).generate(all_words_string)

# Plot the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
# prompt: Vectorizer-> count ->

from sklearn.feature_extraction.text import CountVectorizer

# Join the lemmatized tokens back into strings for vectorization
df['Reviews_lemmatized_str'] = df['Reviews_lemmatized'].apply(lambda tokens: ' '.join(tokens) if tokens is not None else '')

# Initialize the CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the lemmatized review strings
count_matrix = count_vectorizer.fit_transform(df['Reviews_lemmatized_str'])

print("\nShape of the Count Matrix (samples, features):", count_matrix.shape)

# You can optionally view the vocabulary
# print("\nVocabulary (mapping from term to feature index):")
# print(count_vectorizer.vocabulary_)

# To see the count matrix as a dense array (be cautious with large datasets)
# print("\nCount Matrix as Dense Array (first 5 rows):")
# print(count_matrix.toarray()[:5])
```

```python
# prompt: create ML Models any 3

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report

# Define target variable (using 'Stars' as an example, might need transformation for classification)
# For simplicity, let's create a binary target: 1 if Stars > 3, 0 otherwise
df['Sentiment'] = df['Stars'].apply(lambda x: 1 if x > 3 else 0)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(count_matrix, df['Sentiment'], test_size=0.25, random_state=42)

# --- Model 1: Logistic Regression ---
print("--- Training Logistic Regression Model ---")
log_reg = LogisticRegression(max_iter=1000) # Increased max_iter for convergence
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

print("\nLogistic Regression Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Classification Report:\n", classification_report(y_test, y_pred_lr))

# --- Model 2: Multinomial Naive Bayes ---
print("\n--- Training Multinomial Naive Bayes Model ---")
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)

print("\nMultinomial Naive Bayes Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print("Classification Report:\n", classification_report(y_test, y_pred_nb))

# --- Model 3: Linear Support Vector Machine (SVM) ---
print("\n--- Training Linear SVM Model ---")
svm_model = LinearSVC(max_iter=10000) # Increased max_iter for convergence
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print("\nLinear SVM Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Classification Report:\n", classification_report(y_test, y_pred_svm))
```

```python
# prompt: Confusion matrix -> highest

from sklearn.metrics import confusion_matrix

# Confusion Matrix for Logistic Regression
conf_matrix_lr = confusion_matrix(y_test, y_pred_lr)
print("\nConfusion Matrix (Logistic Regression):")
print(conf_matrix_lr)

# Plot Confusion Matrix for Logistic Regression
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_lr, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()

# Confusion Matrix for Multinomial Naive Bayes
conf_matrix_nb = confusion_matrix(y_test, y_pred_nb)
print("\nConfusion Matrix (Multinomial Naive Bayes):")
print(conf_matrix_nb)

# Plot Confusion Matrix for Multinomial Naive Bayes
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nb, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Multinomial Naive Bayes')
plt.show()


# Confusion Matrix for Linear SVM
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
print("\nConfusion Matrix (Linear SVM):")
print(conf_matrix_svm)

# Plot Confusion Matrix for Linear SVM
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_svm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Linear SVM')
plt.show()
```

```python
# prompt: Deploy with gradio Test with 5 sentences

!pip install -q gradio
import gradio as gr

# Define a function that takes a list of sentences and returns predictions
def predict_sentiment(sentences):
  # Ensure sentences is a list
  if not isinstance(sentences, list):
    sentences = [sentences]

  # Preprocess the input sentences
  cleaned_sentences = [remove_punctuation(s.lower().replace(r'\s+', '', regex=True)) for s in sentences]
  tokenized_sentences = [tokenize_text(s) for s in cleaned_sentences]
  cleaned_stopwords_sentences = [remove_stopwords(tokens) for tokens in tokenized_sentences]
  lemmatized_sentences = [lemmatize_tokens(tokens) for tokens in cleaned_stopwords_sentences]
  lemmatized_sentences_str = [' '.join(tokens) if tokens is not None else '' for tokens in lemmatized_sentences]

  # Vectorize the processed sentences using the trained CountVectorizer
  # Use transform here as we already fitted on the training data
  input_count_matrix = count_vectorizer.transform(lemmatized_sentences_str)

  # Make predictions using one of the trained models (e.g., Logistic Regression)
  predictions = log_reg.predict(input_count_matrix)

  # Convert predictions to human-readable labels
  sentiment_labels = ["Negative" if p == 0 else "Positive" for p in predictions]

  # Return the predictions
  return sentiment_labels

# Define the Gradio interface
interface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=5, placeholder="Enter sentences here (one per line)"),
    outputs="text", # Output will be a string showing predictions
    title="Bank Review Sentiment Analysis",
    description="Enter up to 5 bank review sentences to get sentiment predictions (Positive/Negative)."
)

# Launch the Gradio interface
interface.launch()
```
